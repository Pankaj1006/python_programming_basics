## Machine Learning

Machine learning is an application of artificial intelligence (AI) that provides systems the ability to automatically learn and improve from experience without being explicitly programmed.



* **Traditional Software Development** - We Create Programs which takes the input and process them according to the instruction provided by us.
* **Machine Learning** - We create the models which can be trained to perform intelliegent task with computers high computation power


Main areas where we can see Application of Machine learning

* **Predictive Modeling** - We predict about the unkown data with the help of the data we have (like predicting the weather using the old traits.  
* **Computer Vision** - The use of machine learning to detect patterns in visual data. (Face recognition)
* **Time Series** - The use of machine learning to detect patterns in data whuich varies with time. (Speech recognition) 


## Deep Learning

Deep learning is the part of Machine Learning which is based on Neural network. For building the Models in Deep learning we use artificail neural network is used. we can define a deep neural network as neural network with a many layers.

## Artificial Neural Network

Artificial neural networks (ANN) or connectionist systems are computing systems that are inspired by, but not necessarily identical to, the biological neural networks that constitute animal brains. Such systems "learn" to perform tasks by considering examples, generally without being programmed with any task-specific rules. 






## Why Deep Learning

Deep learning is not a new concept. But from years this concept is ignored just becuase of lack of high computation power and popularity of other models. Its the third rise of the ANN but now they are becoming more and more popular because of todays avilability of high power computing resources Like GPU and CPU.





For predictive modeling neural networks are not that different than other models, such as:

    Support Vector Machines
    Random Forests
    Gradient Boosted Machines

Like all other Models and Algorithms the ANN can be used for Classification and Regression models and can produce better results for many tasks.

the two additional advantages of using the ANN the :
* **Convolutional Neural Networks** - Can scan an image for patterns within the image.
* **Recurrent Neural Networks** - Can find patterns across several inputs, not just within a single input.
