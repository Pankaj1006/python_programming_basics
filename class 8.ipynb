# Training and Validation



In order to check the accuracy of the model we have to split the dataset or the training dat ainto test set and validation set.

* **Training Data** - - The data that the machine learning model was fit to/created from. 
* **Validation Data** - - The data that the machine learning model is evaluated upon after it is fit to the training data.


```python
import os
import pandas as pd
import numpy as np


path = '../data/'

read_csv = os.path.join(path , 'auto-mpg.csv')
df = pd.read_csv(read_csv , na_values = ['NA' , '?'])

df = df.reindex(np.random.permutation(df.index))  # reindexing the data set to avoid the overfitting 
df[0:5]
```

# Apply and Map

```python
import os
import pandas as pd
import numpy as np

path = './data/'
read_csv = os.path.join(path, "auto-mpg.csv")
df = pd.read_csv(read_csv,na_values=['NA','?'])

display(df[0:10])
```

```python
df['origin_name'] = df['origin'].map({1: 'North America', 2: 'Europe', 3: 'Asia'})
display(df)
```

```python
effi = df.apply(lambda x: x['displacement']/x['horsepower'], axis=1)
display(effi[0:10])
```
